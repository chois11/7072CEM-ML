# A wine price prediction based on Machine Learning

```python
#Load required libraries

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
```

```python
#Load Dataset to Pandas dataframe
df_raw = pd.read_csv ('./archive/winemag-data-130k-v2.csv')
df_raw.drop("Unnamed: 0",axis=1,inplace=True) # drop factor from axis 1 and make changes permanent by inplace=True
df_raw
```

The purpose of this project is prediect a wine review point so the project will be use Regression.

Cetain columns are unnessary for natural language processing:

To clean dataset,taster_twitter_handle column will be deleted

Training dataset and test dataset are needed for machine learning

```python
#Dropping out the columns
df = df_raw.drop(['description','taster_twitter_handle'], axis=1)
df
```

```python
#Show a number of null or NaN values
df.isnull().sum()
```

```python
# Replace missing values with a string
df['country'].fillna('Other', inplace=True)
df['designation'].fillna('Other', inplace=True)
df['price'].fillna(0, inplace=True)
df['variety'].fillna('Other', inplace=True)
df['province'].fillna('Other', inplace=True)
df['region_1'].fillna('Other', inplace=True)
df['region_2'].fillna('Other', inplace=True)
df['taster_name'].fillna('Other', inplace=True)

df.isnull().sum() # Show the remain NaN values
```

```python
# creating instance of labelencoder
labelencoder = LabelEncoder()
```

```python
dataset_labeled = df.copy()
```

```python
# Assigning numerical values and storing in another column
dataset_labeled['country_Cat'] = labelencoder.fit_transform(dataset_labeled['country'])
dataset_labeled['designation_Cat'] = labelencoder.fit_transform(dataset_labeled['designation'])
dataset_labeled['province_Cat'] = labelencoder.fit_transform(dataset_labeled['province'])
dataset_labeled['region_1_Cat'] = labelencoder.fit_transform(dataset_labeled['region_1'])
dataset_labeled['region_2_Cat'] = labelencoder.fit_transform(dataset_labeled['region_2'])
dataset_labeled['taster_name_Cat'] = labelencoder.fit_transform(dataset_labeled['taster_name'])
dataset_labeled['title_Cat'] = labelencoder.fit_transform(dataset_labeled['title'])
dataset_labeled['variety_Cat'] = labelencoder.fit_transform(dataset_labeled['variety'])
dataset_labeled['winery_Cat'] = labelencoder.fit_transform(dataset_labeled['winery'])
dataset_labeled
```

```python
dataset_labeled = dataset_labeled.drop(['country','designation','province','region_1','region_2','taster_name','title','variety','winery'], axis=1)
```
