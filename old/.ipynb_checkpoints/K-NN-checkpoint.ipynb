# K-Nearest Neighbors Regresstion

```python
# Load required libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

import seaborn as sns
sns.set_theme(style="ticks", color_codes=True)
```

```python
dataset = pd.read_csv('winemag_labeled.csv')
dataset.drop(['title_Cat'],axis=1)
```

### Check dataset data type

```python
# 129971 rows and 10 feature inputs and 1 target output
dataset.shape
```

```python
dataset_train = dataset.copy()
```

```python
dataset.dtypes
```

```python
dataset
```

### Spliting dataset into train and test

```python
X = dataset_train.drop('price', axis=1)
y = dataset_train['price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
X
```

```python
y
```

### K-NN 

```python
# Construct pipeline
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
```

```python
knnregr= Pipeline([
     ('scaler', StandardScaler()),
     ('normalize', Normalizer()),
     ('regressor' , KNeighborsRegressor())
      ])

knnregr.fit(X_train, y_train)
y_pred = knnregr.predict(X_test)

# print(classifier)
knnregr.get_params().keys()
# print("K-NN model score:", knnregr.score(X_test, y_test))



# print(classifier)
print("K-NN model score::", knnregr.score(X_test, y_test))

from sklearn import metrics
print('Mean Absolute Error\n:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error\n:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

### Decision Tree Regression

```python
# Construct pipeline
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
```

```python
# Spliting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
dt = Pipeline([
     ('normalize', Normalizer()),
     ('regressor' , DecisionTreeRegressor()),
])
dt.steps

dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)

# print(classifier)
print("Decision Tree Regression model score:", dt.score(X_test, y_test))
```

```python
from sklearn.model_selection import cross_validate

scores = cross_validate(dt, X_train, y_train)
scores
```

```python
scores['test_score'].mean()
```
