```python
# Load essentail libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


# Make numpy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
 
print(tf.__version__)
```

```python
raw_data = pd.read_csv('winemag_labeled.csv')
raw_data
```

```python
train_dataset = raw_data.sample(frac=0.8, random_state=0)
test_dataset = raw_data.drop(train_dataset.index)
```

```python
train_dataset.describe().transpose()
```

```python
train_features = train_dataset.copy()
test_features = test_dataset.copy()

train_labels = train_features.pop('price_Cat')
test_labels = test_features.pop('price_Cat')
```

```python
train_dataset.describe().transpose()[['mean', 'std']]
```

```python
normalizer = preprocessing.Normalization()
```

```python
normalizer.adapt(np.array(train_features))
```

```python
print(normalizer.mean.numpy())
```

```python
first = np.array(train_features[:1])

with np.printoptions(precision=2, suppress=True):
  print('First example:', first)
  print()
  print('Normalized:', normalizer(first).numpy())
```

```python
points_Cat = np.array(train_features['points_Cat'])

points_Cat_normalizer = preprocessing.Normalization(input_shape=[1,])
points_Cat_normalizer.adapt(points_Cat)
```

```python
points_Cat_model = tf.keras.Sequential([
    points_Cat_normalizer,
    layers.Dense(units=1)
])

points_Cat_model.summary()
```

```python
points_Cat_model.predict(points_Cat[:10])
```

```python
points_Cat_model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
%%time
history = points_Cat_model.fit(
    train_features['points_Cat'], train_labels,
    epochs=250,
    # suppress logging
    verbose=0,
    # Calculate validation results on 20% of the training data
    validation_split = 0.2)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0, 20])
  plt.xlabel('Epoch')
  plt.ylabel('Error [Point]')
  plt.legend()
  plt.grid(True)
```

```python
plot_loss(history)
```
